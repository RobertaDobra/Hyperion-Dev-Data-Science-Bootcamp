##### Using Group-by’s and Merges
Group-by’s can be used to build groups of rows based on a specific feature in your dataset eg. the ‘Manufacturer’ categorical column. 

```python
group_by_Manufacturer = car_data.groupby(by=['Manufacturer'])

car_data_avg = round(group_by_Manufacturer.mean(),0) #0 is number of decimal points
car_data_avg
```

```python
import seaborn as sns
import pandas as pd
df = pd.read_csv('Cars93.csv')
sns.boxplot(data=df, x="Rev.per.mile", y="Manufacturer", order=["Audi", "Hyundai", "Suzuki", "Toyota"]) #x is x-axis, y is y-axis
print("Suzuki has the highest Revs per mile, the only value also being the highest of all, near 3400 rev.per.mile")
```

```python
import matplotlib.pyplot as plt
plt.hist(df['MPG.city'], edgecolor="black", label='MPG.city', alpha=0.5) #alpha is a measure of how transparent the graphs are, the lower the value, the more transparent
plt.hist(df['MPG.highway'], edgecolor="black", label='MPG.highway', alpha=0.5)
plt.legend()
plt.xlabel("MPG")
plt.ylabel("Count")
plt.show()
print("Car can travel more MPG at high frequencies on the highway. Generally it is more fuel efficient to drive on the highway.")
```

```python
sns.lineplot(data=df, x="Wheelbase", y="Turn.circle")
print("The bigger the wheelbase, the bigger the turning circle.")
```

```python
df = pd.read_csv('Cars93.csv')
df.groupby('Type')['Horsepower'].mean() #average in each group
p = sns.barplot(x='Type', y='Horsepower', data=df, ci=False) #ci=confidence interval
print("Larger can has the highest value of horsepower. Can types from small, compact, midsize, large show, on average, increasing horsepower in this order.")
```

```python
#pie chart that has slice for each engine size, proportion determined as count of engines of same size/total of engines
value=df['EngineSize'].value_counts() #counts of unique values
label=df['EngineSize'].unique().tolist() #get unique values in a list
plt.figure(figsize=(10,10))
plt.pie(value, labels=label)
plt.show()
print("Engine of size 1.8 is the most popular i.e has the largest slice in the pie chart")
```
