```python
import pandas as pd
df = df = pd.read_csv('wine.csv') #open dataframe
df.drop_duplicates() #drop duplicate rows from dataframe
df.loc[df.isna().any(axis=1)] #show rows with null values
```

```python
df.info()
```

```python
df.isna().sum() #sum of null values for each column
```

```python
df = df.dropna(subset=['variety'])
```

```python
list=['Cabernet Sauvignon', 'Pinot Noir', 'Chardonnay']
df = df[df.variety.isin(list)] #create dataframe with rows that contain one of the item in the list
df #show dataframe
```

```python
df = df.apply(pd.to_numeric,errors="ignore") #convert all columns to numeric
```

```python
df.dtypes
```

```python
df['points'] = df['points'].astype(float) #convert column to float type
```

```python
df.dtypes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure()
fg=sns.FacetGrid(df, col='variety', margin_titles=True) #3 columns, 3 graphs 
fg.map(sns.histplot,'points') #for each graph a histplot of points is made
fg.add_legend()
plt.show()
```

```python
from wordcloud import WordCloud
df=df.astype({'province':'string', 'variety':'string'}) #convert province and variety columns to string type
```

```python
text = " ".join(i for i in df.province)
# Create and generate a word cloud image:
wordcloud = WordCloud(background_color='white').generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
text = " ".join(i for i in df.variety)
# Create and generate a word cloud image:
wordcloud = WordCloud(background_color='white').generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
