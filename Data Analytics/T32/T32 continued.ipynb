```python
# From the countries dataset scale the "EG.ELC.ACCS.ZS" column and visualise it as well
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('countries.csv')
```

```python
from mlxtend.preprocessing import minmax_scaling

# scale the goals from 0 to 1
scaled_data = minmax_scaling(df['EG.ELC.ACCS.ZS'], columns = [0])

# plot the original & scaled data together to compare
fig, ax=plt.subplots(1,2)
sns.distplot(df['EG.ELC.ACCS.ZS'], ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(scaled_data, ax=ax[1])
ax[1].set_title("Scaled data")
```

```python
from scipy import stats
positive_pledges = df['SP.DYN.CBRT.IN']
normalised_pledges = stats.boxcox(positive_pledges)[0]
```

```python
# From the countries dataset standardise the "SP.DYN.CBRT.IN" column

fig, ax=plt.subplots(1,2)
sns.distplot(positive_pledges, ax=ax[0])
ax[0].set_title("Original data")
sns.distplot(normalised_pledges, ax=ax[1]) #data follows a normal distribution
ax[1].set_title("Normalised data")
```
