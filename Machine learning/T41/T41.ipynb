```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df=pd.read_csv("insurance.csv")
```

```python
x = df.iloc[:,:1].values #all rows from first column
y = df.iloc[:,-1].values #all rows from last column

plt.scatter(x,y,color = 'b')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.show()
```

```python
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.ylabel('charges')
plt.xlabel('age')
plt.show()
```

```python
unk_x = [[27]]

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()

print("For age 27, insurance charges are:", model.predict(unk_x))
```

```python
x = df.iloc[:,2:3].values #all rows from third column
```

```python
plt.scatter(x,y,color = 'b')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.show()
```

```python
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.ylabel('charges')
plt.xlabel('bmi')
plt.show()
```

```python
unk_x = [[27]] 

x_pred = np.append(x, unk_x).reshape(-1,1) 
y_pred = model.predict(x_pred) #fit new model using x and unk_x data

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('bmi')
plt.ylabel('charges')
plt.show()

print("For bmi of 27, insurance charges are:", model.predict(unk_x)) #predict y value for this x value using the new fitted model
```
