```python
# Assigning values to variables 
x = df.iloc[:,[1,4]].values
y = df.iloc[:,5].values
```

```python
# Importing necessary libraries
from sklearn.neighbors import KNeighborsClassifier

# Creating the k-NN classifier
knn = KNeighborsClassifier(n_neighbors=3)

# Training the classifier
knn.fit(x, y)

# Predicting on the testing data
y_pred = knn.predict(x)

# Evaluating the accuracy of the classifier
accuracy = knn.score(x, y)
print("Accuracy:", accuracy)
```

```python
from sklearn.preprocessing import LabelEncoder

# Converting the categorical target variable to numerical values
le = LabelEncoder()
y_pred_num = le.fit_transform(y_pred)
```

```python
import matplotlib.pyplot as plt
```

```python
# Creating a scatter plot of the data, colored by predicted class labels
plt.scatter(x[:, 0], x[:, 1], c=y_pred_num)
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalWidthCm')
plt.title('k-NN Clustering of My Dataset')
plt.show()
```
