```python
import pandas as pd
df = pd.read_csv('automobile.txt')
```

```python
df=df.drop_duplicates()
```

```python
# Replace question mark values with null values
import numpy as np
df=df.replace("?", np.NaN)
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns='normalized-losses')
```

```python
df['make'].value_counts()
```

```python
df['make'].loc[df.isna().any(axis=1)].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data=df['make'].value_counts()
labels=df.make.unique()
plt.figure(figsize=(8,8))
plt.pie(data, labels = labels)
plt.show()
```

```python
#bar chart display highest, lowest average price
df1=df.dropna()
df1 = df1.astype({'price':'int'})
df1.groupby('make')['price'].mean().sort_values().plot.bar(title="Average price per make of automobile")
```

```python
df1 = df1.apply(pd.to_numeric,errors="ignore")
```

```python
df1.plot.scatter(x="bore" ,y="stroke")
```

```python
df1.plot.scatter(x="horsepower", y="peak-rpm")
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=df1["engine-size"],y=df1["city-mpg"],label="City MPG")
plt.scatter(x=df1["engine-size"],y=df1["highway-mpg"],label="Highway MPG")
plt.legend()
plt.xlabel("engine-size")
plt.ylabel("mpg")
```
