```python
import pandas as pd

# Load data
df = pd.read_csv('Iris.csv')

# Select useful features
x = df.iloc[:,[2,3]].values
y = df.iloc[:,5].values
```

```python
from sklearn.preprocessing import MinMaxScaler
# define min max scaler
normalised = MinMaxScaler()
# transform data
x_normalised = normalised.fit_transform(x)                    
```

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

K = 3
plt.figure(figsize=[5,5])

cluster = AgglomerativeClustering(n_clusters=K, affinity='euclidean', linkage='complete')

# Fit the model and obtain predicted cluster labels
cluster_labels = cluster.fit_predict(x_normalised)

# Plot the clusters
plt.scatter(x_normalised[:, 0], x_normalised[:, 1], c=cluster_labels)
plt.xlabel('SepalWidthCm')
plt.ylabel('PetalLengthCm')
plt.show()
```

```python
K = 3
plt.figure(figsize=[5,5])

cluster = AgglomerativeClustering(n_clusters=K, affinity='euclidean', linkage='single')

# Fit the model and obtain predicted cluster labels
cluster_labels = cluster.fit_predict(x_normalised)

# Plot the clusters
plt.scatter(x_normalised[:, 0], x_normalised[:, 1], c=cluster_labels)
plt.xlabel('SepalWidthCm')
plt.ylabel('PetalLengthCm')
plt.show()
```
